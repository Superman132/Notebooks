```
import numpy as np
import matplotlib.pyplot as plt
import time

temps = np.sin(np.linspace(0, 10, num=50))
a = 4.2 # thermal conductivity coefficient
dt = 0.05
n = 500

def heat_equation(points):
  diff = np.ediff1d(points)
  padded = np.insert(diff, 0, 0, axis=0)
  padded = np.insert(padded, points.shape[0], 0, axis=0)
  laplace = np.ediff1d(padded)
  heat_time_partial = a * dt * laplace
  return points + heat_time_partial

for x in range(n):
  temps = heat_equation(temps)
  # print(points)
  if x % int(n/5) == 0:
    plt.plot(temps)
```
